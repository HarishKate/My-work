```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
missing_values=["n/a","NA","N/A","na","--","."]
hk=pd.read_csv("C:\\Users\\Harish Kate\\Downloads\\loans data.csv",na_values=missing_values)
```

```python
hk
```

```python
hk.isna().sum()
```

```python
hk['Employment.Length'].fillna(hk['Employment.Length'].mode().iloc[0], inplace=True)
hk['Loan.Length'].fillna(hk['Loan.Length'].mode().iloc[0], inplace=True)
hk['Loan.Purpose'].fillna(hk['Loan.Purpose'].mode().iloc[0], inplace=True)
hk['Home.Ownership'].fillna(hk['Home.Ownership'].mode().iloc[0], inplace=True)
hk.isnull().sum()
```

```python
hk.drop(columns=['ID'], inplace= True)
hk
```

```python
hk.info()
```

```python
hk= pd.get_dummies(hk,columns=['Loan.Length','Loan.Purpose','Home.Ownership','Employment.Length'],drop_first=True)
```

```python
hk
```

```python
y = hk['Interest.Rate']
x = hk.drop('Interest.Rate',axis= 1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 ,random_state = 69)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model_LR= LinearRegression()
model_LR.fit(x_train,y_train)
```

model_LR.intercept_

```python
model_LR.intercept_
```

```python
model_LR.coef_
```

```python
x_train.columns
```

```python
predictors=x_train.columns
coef=pd.Series(np.reshape(model_LR.coef_,38),predictors).sort_values()
coef.plot(kind='bar',title='Model coefficients',figsize=(15,10))
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
pred = model_LR.predict(x_test)
score_test=r2_score(y_test,pred)
mse_test=mean_squared_error(y_test,pred)
mape_test=mean_absolute_percentage_error(y_test,pred)

print("Test R2 score is:",score_test)
print("Test MSE is:", mse_test)
print("Test MAPE is:", mape_test)
```

```python
pred = model_LR.predict(x_train)
score_train=r2_score(y_train,pred)
mse_train=mean_squared_error(y_train,pred)
mape_train=mean_absolute_percentage_error(y_train,pred)

print("Test R2 score is:",score_train)
print("Test MSE is:", mse_train)
print("Test MAPE is:", mape_train)
```
