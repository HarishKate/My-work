```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv("C:\\Users\\Harish Kate\\Documents\\E Commerce Dataset.csv")
data
```

```python
data.info()
data.isna().sum()
```

```python
data['PreferredPaymentMode']=np.where(data['PreferredPaymentMode']=='CC','Credit Card',data['PreferredPaymentMode'])
data['PreferredPaymentMode']=np.where(data['PreferredPaymentMode']=='COD','Cash on Delivery',data['PreferredPaymentMode'])
```

```python
data
```

```python
data['Tenure'].fillna(data['Tenure'].mode().iloc[0], inplace=True)
data['WarehouseToHome'].fillna(data['WarehouseToHome'].mode().iloc[0], inplace=True)
data['HourSpendOnApp'].fillna(data['HourSpendOnApp'].mode().iloc[0], inplace=True)
data['OrderAmountHikeFromlastYear'].fillna(data['OrderAmountHikeFromlastYear'].mode().iloc[0], inplace=True)
data['CouponUsed'].fillna(data['CouponUsed'].mode().iloc[0], inplace=True)
data['OrderCount'].fillna(data['OrderCount'].mode().iloc[0], inplace=True)
data['DaySinceLastOrder'].fillna(data['DaySinceLastOrder'].mode().iloc[0], inplace=True)
data.isnull().sum()
```

```python
data.drop(columns=['CustomerID'], inplace= True)
data
```

```python
y = data['Churn']
x = data.drop('Churn',axis= 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2 ,random_state = 69)
```
